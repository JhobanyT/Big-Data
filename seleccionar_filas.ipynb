```python
import numpy as np
import pandas as pd
```

```python
url = url = ("https://raw.githubusercontent.com/JhobanyT/Big-Data/main/marketing_data.csv")
df = pd.read_csv(url)
print(df.shape)
df.head()
```

# Seleccionar filas

```python
# seleccion de las filas de la 10 a la 20. 10 inclusiva y la 20 no incluye

df.iloc[10:20]
```

```python
# para saber el nombre de cada una de las columnas
df.columns.values
```

# OTRO TIPO DE SELECCION DE FILAS

```python
# LOS 5 CLIENTES MAS MAYORES
df.nsmallest(5, 'Year_Birth')
```

```python
# LOS 5 CLIENTES MAS JOVENES
df.nlargest(5, 'Year_Birth')
```

```python
# CLIENTES CON MAYOR CANTIDAD DE DIAS DESPUES DE SU ULTIMA COMPRA
df.nlargest(5, "Recency")
```

```python
# CLIENTES CON MENOR CANTIDAD DE DIAS DESPUES DE SU ULTIMA COMPRA
df.nsmallest(5, "Recency")
```
